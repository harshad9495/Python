```python
import pandas as pd
```

```python
# Load the Excel file
df = pd.read_excel('supermarket_sales.xlsx')
```

```python
# Display first few rows
df.head()
```

```python
# Get basic information about the dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics of the dataset
df.describe()
```

```python
# Group by product line and calculate total sales
total_sales_product = df.groupby('Product line')['Total'].sum().reset_index()

# Display the results
total_sales_product
```

```python
# Calculate the average sales amount per transaction
average_sales = df['Total'].mean()

# Display the result
print("\nAverage sales per transaction: ", average_sales)
```

```python
# Group by city and calculate total sales
sales_by_city = df.groupby('City')['Total'].sum().reset_index()

# Display the results
sales_by_city
```

```python
# Count the number of transactions for each payment method
payment_method_count = df['Payment'].value_counts().reset_index()

# Rename the columns for clarity
payment_method_count.columns = ['Payment Method', 'Count']

# Display the result
payment_method_count
```

```python
## Count transactions by gender
gender_count = df['Gender'].value_counts().reset_index()

# Rename the columnsnn   vyt
gender_count.columns = ['Gender', 'Count']

# Display the result
gender_count
```
